### 导包

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
from matplotlib.pyplot import MultipleLocator
from matplotlib import ticker
```

### 设置字体字号

```python
config = {
    "font.family":'serif',
    "font.size": 10.5,
    "mathtext.fontset":'stix',
    "font.serif": ['SimSun'],
}
rcParams.update(config)
```

### 准备数据

```python
y1 = [53.2, 56, 60.4, 64, 66.4, 76.2, 84.8, 85.4, 77, 63.3, 45.8, 56.1, 71.5, 67.6, 62.7, 61.3, 61.8, 67.9, 72.5, 72.5, 69.8, 62.2, 50.9, 46.7]

y1 = [i/100 for i in y1]

y2 = [1-i for i in y1]

x = [i for i in range(24)]
```

### 开始绘图

```python
plt.figure(figsize=(7,4),dpi=600)
```

### 折线图

```python
# 简写
# plt.plot(x,y1,'ob-',label='堂食占比(%)')
plt.plot(x,y1,color='tab:blue',linewidth=2,marker='o',markersize=5,label='堂食占比(%)')
plt.plot(x,y2,label='外卖占比(%)')
# 刻度字体字号
plt.xticks(fontproperties=' Times New Roman', fontsize=10)
plt.yticks(fontproperties=' Times New Roman', fontsize=10)
# 刻度间隔
gap = MultipleLocator(1)
ax = plt.gca()
ax.xaxis.set_major_locator(gap)
# y轴百分比
ax.yaxis.set_major_formatter(ticker.PercentFormatter(xmax=1,decimals=0))
# 坐标轴范围（小数不显示可调整空白）
plt.xlim(-0.7,24)
plt.ylim(0,1)
# 坐标轴名称
plt.xlabel(u'小时')
plt.ylabel('比例')
# 节点标签
for a, b in zip(x, y1):
    plt.text(a, b+0.03, '%.1f%%' % (b*100),horizontalalignment='center', fontsize=7,fontproperties='Times New Roman')#每个点的数值
for a, b in zip(x, y2):
    plt.text(a, b-0.06, '%.1f%%' % (b*100),horizontalalignment='center', fontsize=7,fontproperties='Times New Roman')#每个点的数值
# 备注
plt.legend()
# 网格线
# plt.grid()
# 自动调整
plt.tight_layout()
```

### 散点图

```python
plt.scatter(x,y1,color='tab:blue',linewidth=2,marker='o',label='堂食占比(%)')
plt.scatter(x,y2)
plt.legend()
```

### 优化问题的搜索空间图

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from matplotlib.pyplot import MultipleLocator  # 坐标轴

fig = plt.figure(figsize=(5.5,4), dpi=300)
ax = Axes3D(fig)
X = np.arange(-100,100)
Y = np.arange(-100, 100)
X, Y = np.meshgrid(X, Y)
Z = X**2 + Y**2
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='rainbow')

cset = ax.contourf(X, Y, Z, zdir='z', offset=-400, cmap='rainbow')
# ax.set_title('F1')
ax.w_xaxis.set_pane_color((0, 0, 0, 0))
ax.w_yaxis.set_pane_color((0, 0, 0, 0))
ax.w_zaxis.set_pane_color((0, 0, 0, 0))

# plt.xticks(fontsize=18)
# plt.yticks(fontsize=18)
x_major_locator = MultipleLocator(50)
y_major_locator = MultipleLocator(50)
z_major_locator = MultipleLocator(5000)
ax.set_xlabel('x2')
ax.set_ylabel('x1')
ax.set_zlabel('F1(x1,x2)',labelpad=8,fontsize=14)
ax = plt.gca()  # ax为两条坐标轴的实例

ax.xaxis.set_major_locator(x_major_locator)
ax.yaxis.set_major_locator(y_major_locator)
ax.zaxis.set_major_locator(z_major_locator)

fig.set_tight_layout(False)
ax.view_init(20, 45)#调整旋转角度
# plt.tight_layout()
fig.show()
```
